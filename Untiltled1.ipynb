**UNIVARIATE ANALYSIS**

```python
count=data.airline.value_counts()
plt.pie(count, labels=count.index, autopct='%1.1f%%')
plt.title('Airlines Distribution')
plt.show()
```

```python
count=data.arrival_time.value_counts()
plt.pie(count, labels=count.index, autopct='%1.1f%%')
plt.title('Distribution of arrival time ')
plt.show()
```

```python
count=data.departure_time.value_counts()
plt.pie(count, labels=count.index, autopct='%1.1f%%')
plt.title('Distribution of departure time')
plt.show()
```

From this chart we can conclude:

*  Most of the flights have arrival time in night and evening.
*  Most of the flights have departure time in morning and evening.

```python
sns.countplot(x='airline', data=data, hue='class')
plt.xlabel('Airlines')
plt.title('Flight Counts by Airline')
plt.show()
```

From this graph we can conclude:

* Vistara has maximum no. of flights operating.
* Spicejet has minimum no. of flights operating.
* Moreover we can also conclude that the flights belong   to Economy Class.

```python
sns.countplot(x='stops',data=data, hue='airline')
plt.xlabel('stops')
plt.title('Flight stopage')
plt.show()
```

From this graph we can conclude:

*   We can see most of flights have one stop.
*   There are less no. flights having zero stop.
*   Here hue represents Airlines where we can clearly see that Vistara and Air_India provides the maximum stops, through which we can relate why maximum no. of flights belong to these airlines.

```python

sns.histplot(data['price'], bins=50, color='blue')
plt.xlabel('price')
plt.title('Distribution of price')
plt.show()
```

From this histogram we can conclude that price of most of the flights range between 4000-7000Rs.

**BIVARIATE ANALYSIS**

```python
sns.lineplot(x='days_left', y='price', data=data)
plt.xlabel('Days Left')
plt.ylabel('Price')
plt.title('Day Left vs Price')
plt.show()
```

From this graph we can conclude that the price of flights goes on increasing as the days left goes decreasing.

```python
sns.barplot(x='airline', y='price', data=data)
plt.xlabel('Airlines')
plt.ylabel('Price')
plt.title('Airlines VS Price')
plt.show()
```

From this graph we can conclude that Vistara and Air_India has maximum price and spicejet even having least no. flights,is still the 3rd most expensive airlines because it has less no. of stoppage.

```python
count=data.source_city.value_counts()
sns.countplot(data=data, x='source_city', hue='destination_city')
plt.xlabel('source city')
plt.title('source city')
plt.show()
```

From this graph we can conclude:

* Delhi is the source city and from here maximum flights take off to Banglore and Mumbai.
* Minimum no. of flights travel to Chennai.
* So, the roots from Delhi to Banglore and Mumbai are congested in terms of airways.
* So, potential airline business will be profitable in these roots.

```python
sns.lineplot(x='departure_time',y='price',data=data)
plt.xlabel('departure time')
plt.ylabel('price')
plt.title('departure time VS price')
plt.show()
```

From this lineplot we can say that:

*   Early Morning,Morning and Evening flights are more expensive then others.
*   It shows that the demand for these schedules are high.
*   Even there are maximum no. of flights in these time schedules still the prices are high which shows maximum customers prefer these schedules to travel.

```python
sns.barplot(data=data, x='stops',y ='duration')
plt.xlabel('Stops')
plt.title('Stops VS Duration')
plt.show()
```

From this graph we can say that more stoppage in the journey leads to larger duration of flight and travel longer distance.

```python
sns.lineplot(x='duration',y='price',data=data)
plt.xlabel('duration')
plt.ylabel('price')
plt.title('Duration VS price')
plt.show()
```

From this graph we can conclude:
*   Flights of longer durantion cost more.
*   It could be because of different stops in one journey and larger distance.

```python
sns.barplot(x='stops', y='price',data=data)
plt.xlabel('stops')
plt.ylabel('price')
plt.title('Stops VS price')
plt.show()
```

From above graph we can see flights having two or more stops are more expensive.

**MULTIVARIATE**
